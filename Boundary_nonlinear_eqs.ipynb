# Ecuaciones no lineales para condiciones de contorno

```python
%matplotlib inline

from __future__ import division

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.pyplot import *
%matplotlib inline
 
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
```

```python
g=9.81
```

```python
d=0.6 # Diameter
```

```python
Q=0.1
```

```python
def newton(f,df,x,n,eps):
    it=0
    while ( np.abs(f(x)) > eps and it<n):
        x=x-f(x)/df(x)
        it=it+1
        print x
```

## Flujo de entrada supercritico en lamina libre

Ecuación no lineal para $\theta_B$, ángulo mojado para flujo crítico:
<br>
<br>
\begin{equation}
\frac{h_U}{d}= \frac{1-\cos (\frac{\theta_B}{2})}{2}+\frac{\theta_B-\sin (\theta_B)}{16\sin(\frac{\theta_B}{2})}
\end{equation}

```python
from IPython.display import Image
Image(filename='Descargas/figure2.png')
```

```python
hu=0.5
def ff(t): return ( hu/d-(1-np.cos(t/2))/2 
                      -(t-np.sin(t))/(16*np.sin(t/2)) )
def dff(t): return (-1/2*np.sin(t/2)*1/2
                    -( (1-np.cos(t))*16*np.sin(t/2) 
                    -(t-np.sin(t))*16*np.cos(t/2)*1/2 )  
                    /( (16*np.sin(t/2))**2 ))
x=2*np.pi*np.linspace(0.0001,0.9999,100)
plt.plot(x,ff(x))
plt.plot(x,0.*x)
axes = plt.gca()
axes.set_xlim([0.,7.])
axes.set_ylim([-10,10])
```

```python
hu=0.5
x=2.*np.pi*0.9999
newton(ff,dff,x,20,1.e-6)
```

## Flujo salida subcritico en lamina libre


Ecuación no lineal para $\theta_c$, ángulo mojado para flujo crítico:
<br>
<br>
\begin{equation}
512\frac{Q^2}{gd^5}=\frac{(\theta_c-\sin \theta_c)^3}{\sin{\frac{\theta_c}{2}}}
\end{equation}




<!--<img src="Descargas/figure.png" style="width: 800px;"> -->

```python
from IPython.display import Image
Image(filename='Descargas/figure.png')
```

<!--![alt text](./Descargas/figure.png)-->

```python
def f(x): return 512.*Q/(g*d**5)-(x-np.sin(x))**3/np.sin(x/2.)
def df(x): return -(3.*(x-np.sin(x))**2*(1-np.cos(x))*np.sin(x/2)-(x-np.sin(x))**3*np.cos(x/2)*1./2 )/np.sin(x/2)**2
```

```python
x=2*np.pi*np.linspace(0.0001,0.9999,100)
```

```python
plt.plot(x,f(x))
plt.plot(x,0.*x)
axes = plt.gca()
axes.set_xlim([0.,7.])
axes.set_ylim([-100,100])
```

```python
x=2.*np.pi*0.9999
print np.abs(df(x/2))
```

```python
newton(f,df,x,20,1.e-6)
```

## Flujo de entrada subcrítico en lámina libre

Ecuación no lineal para $\theta_B$, ángulo mojado para flujo crítico:
<br>
<br>
\begin{equation}
h_U= d\frac{1-\cos (\frac{\theta_B}{2})}{2}+ \frac{1}{2g} \left(u_C-\phi_C +\beta \sqrt{gd/8} \sin (\frac{\theta_B}{4}) \right)^2  \quad (3.55)
\end{equation}

Se debe resolver en el intervalo $[\theta_m,2\pi]$ donde $\theta_m$ es el ángulo mínimo para que se produzca flujo de entrada y se calcula como:
<br>
<br>
\begin{equation}
\theta_m=4\arcsin  \left( \frac{\phi_C-u_C}{\beta \sqrt{\frac{gd}{8}} }   \right)
\end{equation}


```python
from IPython.display import Image
Image(filename='Descargas/figure3.png')
```

```python
uc=1
phic=0.1 # Mirar como se calculan este invariante de Riemann?
beta=6.41

def fh(t): return ( hu -1./2*d*(1-np.cos(t/2)) 
                       - 1./(2*g)*(uc-phic+beta*np.sqrt(g*d/8)*np.sin(t/4))**2 ) 
                       
def dfh(t): return (-1./2*d*np.sin(t/2)*1/2
                    - 1/(2*g)* 2* (uc-phic+beta*np.sqrt(g*d/8)*np.sin(t/4)) 
                    * beta*np.sqrt(g*d/8)*np.cos(t/4)*1/4. )
x=2*np.pi*np.linspace(0.0001,0.9999,100)
plt.plot(x,fh(x))
plt.plot(x,0.*x)
axes = plt.gca()
axes.set_xlim([0.,7.])
axes.set_ylim([-10,10])
```

```python
tmin=2.*np.pi*0.0001 # Calcular tmin con formula (3.56)
x=tmin
newton(fh,dfh,x,20,1.e-6)
```
